```python
# Text classification for Sentimental analysis using KNN. (Refer any dataset like Titanic, Twitter, etc.) 
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = pd.read_csv('yelp.csv.xls')
```

```python
data
```

```python
data['length'] = data['text'].apply(len)
data.head()
```

```python
data_classes = data[(data['stars'] == 1) | (data['stars'] == 3) | (data['stars'] == 5)]

```

```python
x = data_classes['text']
y = data_classes['stars']
```

```python
# classification
vocab = CountVectorizer(stop_words='english')
```

```python
x_transformed = vocab.fit_transform(x)
```

```python
print("Shape of the sparse matrix:", x_transformed.shape)
```

```python
print("Non-zero occurrences:", x_transformed.nnz)
     
```

```python
density = (x_transformed.nnz / (x_transformed.shape[0] * x_transformed.shape[1])) * 100
print("Density of the matrix =", density)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_transformed, y, test_size=0.2, random_state=101)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
```

```python
pred_knn = knn.predict(x_test)
```

```python
print("Confusion Matrix for K Neighbors Classifier:")
print(confusion_matrix(y_test, pred_knn))
print("Score:", round(accuracy_score(y_test, pred_knn) * 100, 2))
```
